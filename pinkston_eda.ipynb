```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
## Data Acquisition

# Load the Iris dataset into DataFrame
df = sns.load_dataset('iris')

# Inspect first rows of the DataFrame
print(df.head())
```

```python
## Initial Data Inspection

print(df.head(10))
print(df.shape)
print(df.dtypes)
```

```python
## Summary Statistics

print(df.describe())
```

```python
## Initial Data Distribution for Numerical Columns

# Inspect histogram by numerical column
df['petal_width'].hist()

# Inspect histograms for all numerical columns
df.hist()

# Show all plots
plt.show()
```

There is an obvious correlation between petal length and width; whereas there is more of a variance between the sepal length and width.

```python
## Initial Data Distribution for Categorical Columns

# Inspect value counts by categorical column
df['species'].value_counts()

# Inspect value counts for all categorical columns
for col in df.select_dtypes(include=['object', 'category']).columns:
    # Display count plot
    sns.countplot(x=col, data=df)
    plt.title(f'Distribution of {col}')
    plt.show()

# Show all plots
plt.show()
```

The count of the three species is equal.

```python
## Initial Data Transformation and Feature Engineering

# Renaming a column
df.rename(columns={'sepal_length': 'Sepal Length'}, inplace=True)

# Adding a new column
df['Sepal Area'] = df['Sepal Length'] * df['sepal_width']
df['petal_area'] = df['petal_length'] * df['petal_width']
```

```python
## Data Visualizations

sns.pairplot(df, hue='species')
plt.show()
```

The versicolor and virginica have many similarities; however, the setosa seems to be outliers to the other two species.

```python
## 2nd Data Visualization

sns.displot(df, x="petal_length", hue="species", multiple="dodge")
plt.show()
```

I like this one. It compares the petal length of each species side-by-side.

```python
## 3rd Data Visualization using Custom Column

sns.displot(df, x="petal_area", hue="species", multiple="dodge")
plt.show()
```

This visualization shows the custom column of "petal_area". It helps emphasize that the setosa species has the largest petals of the three species.
